4. 1 (3 puntos) ¿Cuáles son las 10 películas que contaron con más presupuesto?

```python
# leer el archivo csv y guardar en un dataframe
df = pd.read_csv('movies.csv', encoding= 'unicode_escape')

# ordenar el dataframe por la columna 'budget' en forma descendente
df = df.sort_values(by='budget', ascending=False)

# seleccionar las primeras 10 filas
top_10 = df.head(10)

# imprimir las 10 películas con más presupuesto
print(top_10[['title', 'budget']])
```

4. 5 (8 puntos) ¿Cuántas películas se hicieron en cada año? ¿En qué año se hicieron más películas? Haga un gráfico de barras

```python
# convertir la columna 'releaseDate' a tipo fecha
df['releaseDate'] = pd.to_datetime(df['releaseDate'], format='%Y/%m/%d')

# agrupar por año y contar el número de películas por año
movies_by_year = df.groupby(df['releaseDate'].dt.year).count()

# imprimir el número de películas por año
print(movies_by_year[['title']])

# calcular el año con más películas
most_movies_year = movies_by_year['title'].idxmax()
print(f"Año con más películas: {most_movies_year}")

# grafico de barras
movies_by_year['title'].plot(kind='bar')
plt.xlabel('Año')
plt.ylabel('Número de películas')
plt.title('Número de películas por año')
plt.xticks(rotation=90)
plt.show()
```

4. 6. (9 puntos) ¿Cuál es el género principal de las 20 películas más recientes? ¿Cuál es el género principal que predomina en el conjunto de datos? Represéntelo usando un gráfico.

```python
# convertir la columna 'releaseDate' a tipo fecha
df['releaseDate'] = pd.to_datetime(df['releaseDate'], format='%Y/%m/%d')

prueba = df.sort_values(by='releaseDate', ascending=False)

ultimas20 = prueba[['releaseDate','title','genres']].head(20).dropna()

genres_list = ultimas20['genres'].str.split('|')

# print(ultimas20)

diccionario_generos = {}

for genre in genres_list:
    
    for entry in genre:
        
        diccionario_generos[entry] =  0
        
for genre in genres_list:
    
    for entry in genre:
        
        diccionario_generos[entry] =  diccionario_generos[entry] + 1

plt.bar(diccionario_generos.keys(), diccionario_generos.values())
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Count of lastest 20 Movies by Genre')
plt.xticks(rotation=90)
plt.show()


todas = prueba[['releaseDate','title','genres']].dropna()
genres_list = todas['genres'].str.split('|')
diccionario_generos = {}
for genre in genres_list:
    
    for entry in genre:
        
        diccionario_generos[entry] =  0
        
for genre in genres_list:
    
    for entry in genre:
        
        diccionario_generos[entry] =  diccionario_generos[entry] + 1

plt.bar(diccionario_generos.keys(), diccionario_generos.values())
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Count of all time Movies by Genre')
plt.xticks(rotation=90)
plt.show()
```

4. 8 (3 puntos) ¿La cantidad de actores influye en los ingresos de las películas?¿se han hecho
películas con más actores en los últimos años?

```python
# temp = prueba[['title','actorsAmount','revenue','genres']].dropna()
temp = prueba[['actorsAmount','revenue','genres']].dropna()


temp = temp.sort_values(by='actorsAmount', ascending=False)

# print(temp.head(20))

temp.plot.scatter(x='actorsAmount', y='revenue')
```

Como puede visualizar en la grafica previa, la dispersion no tiene ningun tipo de patron, ni tiende a seguir ningun comportamiento por lo que podemos asumir que la cantidad de actores no influencia directamente en que tantas ganancias tiene una pelicula. 

```python
temp = prueba[['actorsAmount','releaseDate','genres']].dropna().sort_values(by='releaseDate', ascending=False)

# print(temp.head(20).to_string(index=False))

temp['year'] = temp['releaseDate'].dt.year

grouped_df = temp.groupby('year').sum()

plt.bar(grouped_df.index, grouped_df['actorsAmount'])
plt.xlabel('Year')
plt.ylabel('Actors Amount')
plt.title('Actors Amount per Year')
plt.show()
```

Si hubo un incremento en el numero de actores conforme avanzaban las decadas pero en las ultimas entradas pero se puede ver que hay un comportamiento de decremento en los ultimos años desde el 2018.
