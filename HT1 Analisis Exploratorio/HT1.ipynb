## 3. (6 puntos) Investigue si las variables cuantitativas siguen una distribución normal y haga una tabla de frecuencias de las variables cualitativas. Explique todos los resultados. 

```python
df_temporal_inciso3 = pd.read_csv('movies.csv', encoding= 'unicode_escape')

for column in df_temporal_inciso3.columns:
    
    if df_temporal_inciso3[column].dtype != 'object':
        
        print(stats.normaltest(df_temporal_inciso3[column]))
        stat, p = stats.normaltest(df_temporal_inciso3[column])
        
        alpha = 0.05
        if p > alpha:
            print(f"La variable {column} sigue una distribución normal (p = {p:.4f})")
        else:
            print(f"La variable {column} no sigue una distribución normal (p = {p:.4f})")
```

```python
# Seleccionar la variable cualitativa
column = 'genres'

# Obtener la tabla de frecuencias
value_counts = df_temporal_inciso3[column].value_counts()

# Graficar la tabla de frecuencias
value_counts.plot(kind='bar', figsize=(12, 6))
plt.title(f'Tabla de frecuencias para la variable {column}')
plt.xlabel(column)
plt.ylabel('Frecuencia')
plt.show()
```

# 4.13 (8 puntos) ¿En qué meses se han visto los lanzamientos con mejores ingresos?¿cuantas películas, en promedio, se han lanzado por mes?

```python

df_real_month = prueba[['releaseDate','revenue','genres']].dropna()

df_real_month = df_real_month[df_real_month.revenue != 0][['releaseDate']]

grouped = df_real_month.groupby(df_real_month['releaseDate'].dt.year)

# Contar la frecuencia de cada mes
month_counts = {}
for name, group in grouped:
    month_counts[name] = group['releaseDate'].dt.month.value_counts().to_dict()

# Hacer la media de los conteos de meses
total_months = {}
for year, counts in month_counts.items():
    for month, count in counts.items():
        if month not in total_months:
            total_months[month] = count
        else:
            total_months[month] += count

# Dividir la suma de los conteos de cada mes por el número de años
average_months = {month: count / len(month_counts) for month, count in total_months.items()}


# Ordenar el diccionario y graficar los resultados
sorted_months = sorted(average_months.items(), key=lambda x: x[1], reverse=True)
months, counts = zip(*sorted_months)

plt.bar(months, counts)
plt.xlabel('Month')
plt.ylabel('Average Count')
plt.title('Average Count of Months')
plt.show()
    
```

Dado el inciso anterior se puede ver que los meses con mejores ingresos son los meses de vacaciones como junio y diciembre. 

Basado en el grafico de mes contra promedio de lanzamientos, se puede ver que los meses con mayores lanzamientos son septiembre y diciembre. 

# 4. 14. (7 puntos) ¿Cómo se correlacionan las calificaciones con el éxito comercial?

```python
dataframe_temporal = prueba[['title','genres','popularity','voteAvg']].dropna()

dataframe_temporal = prueba[['popularity','voteAvg']]

# ipd.display(dataframe_temporal.head(10))

# print(dataframe_temporal.to_string(index=False))

correlation = dataframe_temporal['voteAvg'].corr(dataframe_temporal['popularity'])
print(f'La correlación entre voteAvg y popularity es {correlation:.2f}')

plt.scatter(dataframe_temporal['voteAvg'], dataframe_temporal['popularity'])
plt.xlabel('Vote Average')
plt.ylabel('Popularity')
plt.title('Relación entre Vote Average y Popularity')
plt.show()


```

El hecho de que la correlación entre voteAvg y popularity sea de 0.07 significa que hay una relación muy débil entre estas dos variables. Una correlación cercana a 0 indica que no hay una relación lineal clara entre las dos variables, y por lo tanto, la calificación de una película no es un predictor confiable de su éxito comercial.

Es importante tener en cuenta que la correlación no es una medida de causalidad, por lo que no podemos afirmar que una alta calificación cause un éxito comercial o viceversa. Es posible que haya otros factores que influyan en el éxito comercial de una película, como el género, el reparto, la publicidad, entre otros.

En resumen, la correlación débil entre voteAvg y popularity sugiere que la calificación de una película no es un factor determinante en su éxito comercial, y que hay otros factores que deben ser considerados para predecir el éxito de una película.

# 4. 15. (5 puntos) ¿A qué género principal pertenecen las películas más largas?

```python
dataframe_temporal = prueba[['runtime','genres']].dropna()

# print(dataframe_temporal.to_string(index=False))

dataframe_temporal['primary_genre'] = dataframe_temporal['genres'].str.split('|').str[0]

# Group the dataframe by primary genre and calculate the mean runtime for each genre
grouped = dataframe_temporal.groupby(by='primary_genre').mean()

# Sort the dataframe by runtime in descending order
grouped = grouped.sort_values(by='runtime', ascending=False)

# Plot the data using a bar chart
grouped.plot(kind='bar', y='runtime', color='blue', legend=False)
plt.xlabel('Primary Genre')
plt.ylabel('Mean Runtime (minutes)')
plt.title('Mean Runtime by Primary Genre')
plt.xticks(rotation=90)
plt.show()
```

Según los datos proporcionados, la conclusión es que el género de historia tiene la duración promedio más alta de todos los géneros considerados. Este resultado sugiere que las películas de historia suelen ser más largas que otras películas de otros géneros. Sin embargo, es importante tener en cuenta que esta conclusión se basa en una muestra pequeña de datos y que los resultados pueden variar si se consideran más películas y géneros adicionales. Por lo tanto, se requiere una investigación más exhaustiva para obtener conclusiones más precisas y fiables sobre las duraciones promedio de los géneros cinematográficos.
